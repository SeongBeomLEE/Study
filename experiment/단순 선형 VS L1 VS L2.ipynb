```python
'''
Y = X * [5, 0, 0, 0, 0, 0]
'''

import numpy as np
np.random.seed(22)
X = np.random.rand(100, 6)
Y = 5 * X[:, 0]
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

lr = LinearRegression(fit_intercept=False)
lr.fit(X, Y)
print(f'W:{lr.coef_}, b:{lr.intercept_}, score: {lr.score(X, Y)}')

import matplotlib.pyplot as plt

plt.scatter(Y, lr.predict(X))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

lr = Lasso(fit_intercept=False)
lr.fit(X, Y)
print(f'W:{lr.coef_}, b:{lr.intercept_}, score: {lr.score(X, Y)}')

import matplotlib.pyplot as plt

plt.scatter(Y, lr.predict(X))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

lr = Ridge(fit_intercept=False)
lr.fit(X, Y)
print(f'W:{lr.coef_}, b:{lr.intercept_}, score: {lr.score(X, Y)}')

import matplotlib.pyplot as plt

plt.scatter(Y, lr.predict(X))
plt.show()
```

변수 선택으로써 L1 규제를 사용하는 것이 무조건 좋은 결과를 가져오는 것은 아님

모델 입장에서는 줄여야 할 loss가 2개가 되기 때문에 에측 자체의 성능이 감소할 수 있음

따라서 각 규제의 특성을 잘 이해하고 이에 맞는 모델을 개발하는 것이 중요함
