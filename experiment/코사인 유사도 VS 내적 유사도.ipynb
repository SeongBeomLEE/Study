```python
import numpy as np

vector = np.array([
                   [-10, -10],
                   [-1, -1],
                    [0.5, 0.5],
                    [1, 1],
                    [100, 100],
                   [0, 100000],
                   ])

norm_vector = vector / ((np.sum(vector ** 2, axis = 1)) ** 0.5).reshape(-1, 1)
```

```python
np.dot(vector, vector.T)
```

```python
np.dot(norm_vector, norm_vector.T)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector, vector)
```

```python
cosine_similarity(norm_vector, norm_vector)
```

벡터를 정규화 하고 내적 유사도를 계산한다면 코사인 유사도와 동일하다는 것을 알 수 있음
