```python
import numpy as np
import pandas as pd

X1 = np.array([i for i in range(-100, 101)])
X2 = X1 * 2
X3 = X2 * 5

df = pd.DataFrame()
df['X1'] = X1
df['X2'] = X2
df['X3'] = X3
df.corr()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca_df = pd.DataFrame(pca.fit_transform(df.values))
pca_df.corr()
```

PCA로 새롭게 구한 주성분은 고유벡터를 기준으로 정사영된 벡터이기 때문에 벡터들간의 선형 독립을 보장함
