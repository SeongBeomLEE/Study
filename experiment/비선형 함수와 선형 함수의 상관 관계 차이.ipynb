```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([i for i in range(-100, 101)])
Y = X**2
```

```python
plt.plot(X, Y)
```

```python
import pandas as pd

df = pd.DataFrame()
df['X'] = X
df['Y'] = Y
df.corr()
```
